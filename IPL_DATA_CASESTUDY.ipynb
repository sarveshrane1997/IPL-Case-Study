```python
#import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#read ipl.csv file and print first 5 records
df=pd.read_csv("ipl data.csv")
df.set_index("id",inplace=True)
df
```

```python
#check total number of columns,entries note down your findings
df.info()
```

<b style="color:red">Insights</b>
<p><b>There are total 636 entries all over and 17 columns<b></p>

```python
# find null values 
df["season"].isnull().sum()
```

<b style="color:red">Insights</b>
<p><b>(1)Some columns are containing null data<br>
(2)Umpire has the highest amount of null data</b></p>

```python
#drop null values of city , winner ,player_of_match,umpire1,umpire2 column
df.dropna(subset=["city","winner","player_of_match","umpire1","umpire2"],inplace=True)
df.isnull().sum()
```

```python
#drop umpire3 column as it contains more than 75% of null values
df.drop("umpire3",axis=1,inplace=True)
df.isnull().sum()
```

```python
#which city hosted most number of matches?
df["city"].value_counts()
```

```python
#draw bar plot and write down your insights
df["city"].value_counts().plot(kind="bar")
```

<b style="color:red">Insights</b>
<p><b>(1)In mumbai most ipl match have been played.<br>(2)In India,less than 10 matches played in nagpur.<br>
(3)On the other side Bloemfontein has the least amount of matches played</b></p>

```python
#find all venue of mumbai city
df[df["city"]=="Mumbai"]["venue"].unique()
```

<b style="color:red">Insights</b>
<p><b>There are total of 3 stadiums in Mumbai</b></p>

```python
#now compare in which venue of mumbai most number of matches played (draw bar plot and  write down insights)
a=df[df["city"]=="Mumbai"]["venue"].unique()
b=[]
for i in a:
    c=df[df["venue"]==i]["venue"].value_counts()
    b.append(c[0])
plt.bar(a,b)
plt.xticks(rotation=10)
plt.show()
```

<b style="color:red">Insights</b>
<p><b>In Mumbai, more than 50 matches played in wankhede stadium.</b></p>

```python
#what is the preferred choice after winning a toss in mumbai
df[df["city"]=="Mumbai"]["toss_decision"].value_counts()
```

<b style="color:red">Insights</b>
<p><b>By analyzing this we can state that after winning toss most teams choose field in Mumbai While very few team considers to choose bat</b></p>

```python
#graphical representation of above question
a=df[df["city"]=="Mumbai"]["toss_decision"].unique()
b=[]
c=df[df["city"]=="Mumbai"]["toss_decision"].value_counts()
b.extend([c[0],c[1]])
plt.bar(a,b)
plt.show()
```

```python
#which team won most number of toss
df["toss_winner"].value_counts()
```

<b style="color:red">Insights</b>
<p><b>Mumbai Indians won the most number of toss followed by Kolkata Night Riders
Rising Pune Super has won least number of toss throughout the other teams</b></p>

```python
#show graphical representation of above question
df["toss_winner"].value_counts().plot(kind="bar")
plt.show()
```

```python
#find what mumbai indians preferred after winning a toss?
df[df["toss_winner"]=="Mumbai Indians"]["toss_decision"].value_counts()
```

<b style="color:red">Insights</b>
<p><b>Mumbai indians mostly prefers to field than batting after winning the toss
However there is a slight difference in chooseing decisions</b></p>

```python
#head to head winning count of Mumbai Indians vs Chennai Super Kings
df[(df["team1"]=="Mumbai Indians")&(df["team2"]=="Chennai Super Kings")|(df["team2"]=="Mumbai Indians")&(df["team1"]=="Chennai Super Kings")]["winner"].value_counts()
```

<b style="color:red">Insights</b>
<p><b>(1)Most matches between Mumbai Indians and Chennai Super Kings are won by Mumbai Indians
<br>(2)There is a chase of winning by Chennai Super Kings over Mumbai Indians too</b></p>

```python
#Which team won most of the matches in mumbai?
df[df["city"]=="Mumbai"]["winner"].value_counts()
```

<b style="color:red">Insights</b>
<p><b>(1)In mumbai, Mumbai Indians has won the most matches throughout all teams

```python
#how many times each team won the toss and won the match
df[df["toss_winner"]==df["winner"]]["winner"].value_counts()
```

<b style="color:red">Insights</b>
<p><b>(1)Mumbai indians have won most tosses also won most matches as compared to other teams
<br>(2)Pune has won the least matches through other teams</b></p>

<b style="color:red">Insights</b>
